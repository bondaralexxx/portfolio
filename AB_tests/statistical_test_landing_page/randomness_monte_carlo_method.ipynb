```python
#  For example, we have two landing pages with different conversion rates of 3% and 5%.

a = np.random.binomial(1,0.03,size=1000).mean()
b = np.random.binomial(1,0.05,size=1000).mean()
```

```python
a, b
```

```python
# Using the Monte Carlo method, we will try to understand: Can the actual conversion on the first page be less than the conversion on the second page?

n = 1000
result = []
for i in range(n):
  a = np.random.binomial(1,0.03,size=1000).mean()
  b = np.random.binomial(1,0.05,size=1000).mean()
  result.append((a,b))
```

```python
df = pd.DataFrame(result,columns=['a','b'])
```

```python
df
```

```python
df[df['a'] > df['b']]
```

```python
# The answer is ""YES"", it may be, and this happens because the sample is small and the role of chance is high.
```
