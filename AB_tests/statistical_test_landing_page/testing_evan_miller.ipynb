```python
import pandas as pd
import numpy as np
```

```python
"https://www.evanmiller.org/ab-testing/sample-size.html"
# Baseline conversion rate: 3%
# MDE: 2%
# power: 80%
# Significance level: 5%
# calculated sample size according to Evan Miller's calculator:
samle_size = 1245
```

```python
def test(conv_a,conv_b,size_a,size_b,significance=0.05):
  _, p_value = proportions_ztest([conv_a, conv_b],
                                 [size_a, size_b],
                                 alternative='two-sided')
  return p_value < significance
```

```python
# TRP (Sensitivity) testing
```

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
n = 10000
result = []
for i in range(n):
  a = np.random.binomial(samle_size,0.03)
  b = np.random.binomial(samle_size,0.05)
  result.append((a,b))
```

```python
df = pd.DataFrame(result, columns=['a','b'])
```

```python
#   table with 10000 experiments
```

```python
df
```

```python
df.apply(lambda row: test(row['a'],row['b'],samle_size,samle_size),axis=1)
```

```python
df['test'] = df.apply(lambda row: test(row['a'],row['b'],samle_size,samle_size),axis=1)
```

```python
df
```

```python
df['test'].mean()
```
