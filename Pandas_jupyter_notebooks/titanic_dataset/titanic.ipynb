```python
#  Grouping (groupby)
```

```python
# calculating the average age for males and females
df.groupby('Sex')['Age'].mean()
```

```python
#  calculation of statistical indicators
df.groupby('Sex')['Age'].describe()
```

```python
# calculating of median and average age for males and females, and for survivors and non-survivors
df.groupby(['Sex', 'Survived'])['Age'].agg(['mean', 'median'])
```
