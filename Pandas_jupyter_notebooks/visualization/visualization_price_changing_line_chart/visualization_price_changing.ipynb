```python
# 1. IMPORTING LIBRARIES.

import pandas as pd
import matplotlib.pyplot as plt
```

```python
ibm = pd.read_csv('IBM.csv', index_col='Date')
ibm.head(3)
```

```python
# 2. PREPARATION.

ibm_close_price = ibm['Close']
ibm_close_price.head(3)
```

```python
# 3. PLOTTING AND SETTING UP A GRAPH.

ibm_close_price.plot(figsize=(9,3), color='red')
plt.title('Price of 1 share at market close')
plt.ylabel('Price, $')
plt.xticks(rotation=45, ha='right')
plt.tick_params(axis='y', labelsize=8)
plt.tick_params(axis='x', labelsize=8)
```
